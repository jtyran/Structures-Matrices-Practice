```python


import pandas as pd

from sklearn.datasets import load_iris

#load the iriis data from sklearn
iris_data = load_iris()

df = pd.DataFrame(data=iris_data.data,
                 columns=iris_data.feature_names)

df.head()
```

```python
# Create a pair plot of some columns
# Relationship b/n the two columns
sns.pairplot(iris.iloc[:,:],hue='variety')
```

```python
# Boxplot
iris.boxplot(by="variety", figsize=(15,15))
```

```python
# Splitting the data into training and testing datasets
X=iris.iloc[:,0:4]
Y=iris["variety"]
X.head()
```

```python
#
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25,random_state=0)
print("Train Shape", X_train.shape)
print("Test Shape", X_test.shape)

# Create a logistic regression model
model_lr = LogisticRegression()

# Train model
model_lr.fit(X_train,Y_train)

# Make predictions on the test set
Y_pred = model_lr.predict(X_test)

# Evaluate the model
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(Y_pred, Y_test))
```

```python
# Generate the confusion matrix
conf_matrix = confusion_matrix(Y_test, Y_pred)
print(conf_matrix)
```

```python
from sklearn.neighbors import KNeighborsClassifier # for K nearest neighbours 
from sklearn.svm import SVC #for Support Vector Machine (SVM) Algorithm
from sklearn.tree import DecisionTreeClassifier #for using Decision Tree Algoithm

# DecisonTree 
tree=DecisionTreeClassifier()
tree.fit(X_train,Y_train)
prediction=tree.predict(X_test)
print('The accuracy of the Decision Tree is',metrics.accuracy_score(prediction,Y_test))
```

```python
# K Nearest Neighbor Classification
# Examines 3 neighbours for putting the new data into a class
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
prediction=knn.predict(X_test)
print('The accuracy of the KNN is',metrics.accuracy_score(prediction,Y_test))
```

```python
# SVM
svc=SVC()
svc.fit(X_train,Y_train)
prediction=svc.predict(X_test)
print('The accuracy of the SVC is',metrics.accuracy_score(prediction,Y_test))
```
